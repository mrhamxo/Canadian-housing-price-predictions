```python
# ── Load Raw Data ─────────────────────────────────
df = pd.read_csv("../dataset/raw/HouseListings-Top45Cities.csv")
df.head(10)
```

```python
print("Shape:", df.shape)
```

```python
# ── Basic Info ────────────────────────────────────
print("── Data Types ──")
df.dtypes
```

```python
print("── Missing Values ──")
df.isnull().sum()
```

```python
print("── Basic Stats ──")
df.describe()
```

```python
# ── checking empty data and filling ─────────────────────
print("── Number_Beds ──")
Number_Beds = df[(df["Number_Beds"] == 0)]
Number_Beds.shape
```

```python
print("── Number_Baths ──")
Number_Baths = df[(df["Number_Baths"] == 0)]
Number_Baths.shape
```

```python
print("── Both Number_Beds and Number_Baths ──")
Both_Beds_Baths = df[((df["Number_Beds"] == 0) & (df["Number_Baths"] == 0))]
Both_Beds_Baths.shape
```

```python
df.shape
```

```python
# Replace 0 with NaN first
df["Number_Beds"]  = df["Number_Beds"].replace(0, np.nan)
df["Number_Baths"] = df["Number_Baths"].replace(0, np.nan)
```

```python
# Fill using median per city
df["Number_Beds"]  = df.groupby("City")["Number_Beds"].transform(
    lambda x: x.fillna(x.median())
)
df["Number_Baths"] = df.groupby("City")["Number_Baths"].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.drop(columns=["Address"], inplace=True)
```

```python
# ── Remove Duplicates ─────────────────────────────
print(df.duplicated().sum())
```

```python
df[df.duplicated()].head()
```

```python
before = len(df)
df.drop_duplicates(inplace=True)
print(f"Dropped {before - len(df)} duplicates. Remaining: {len(df)}")
```

```python
# ── Strip Whitespace from Strings ────────────────
df.apply(lambda col: col.astype(str).str.contains(r'^\s|\s$', regex=True).sum())
```

```python
# ── Validate Lat / Lon ────────────────────────────
print("\nLatitude  — min:", df["Latitude"].min(),  "max:", df["Latitude"].max())
print("Longitude — min:", df["Longitude"].min(), "max:", df["Longitude"].max())
print("Missing coords:", df[["Latitude","Longitude"]].isnull().sum().sum())
```

```python
# Save cleaned dataset
df.to_csv("../dataset/cleaned/housing_cleaned.csv", index=False)
print("\n✅ Saved → data/processed/housing_cleaned.csv")
```
